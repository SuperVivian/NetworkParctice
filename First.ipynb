```python
import networkx as nx
G = nx.Graph()#创建一个空图
```

```python
#图是节点和节点对的集合。
#节点是任何可以散列的对象。
G
```

```python
#一次添加一个节点
G.add_node(1)
```

```python
#从可迭代容器中添加节点
G.add_nodes_from([2,3])
```

```python
#查看图的信息
print(G.nodes)
print(G.edges)
print(G.adj)
print(G.degree)
```

```python
H = nx.path_graph(10)
print(H.nodes)
```

```python
G.add_node(H)
print(G.nodes)
```

```python
#一次添加一条边
G.add_edge(1,2)
print(G.edges)
```

```python
#用元组的形式添加边
e = (2,3)
G.add_edge(*e)
print(G.edges)
```

```python
#可视化当前的图
#NetworkX 主要不是图形绘图包，而是包含使用 Matplotlib 进行基本绘图以及使用开源 Graphviz 软件包的接口。
#这些是networkx.drawing模块的一部分，如果可能，将被导入。
import matplotlib.pyplot as plt
%matplotlib inline

nx.draw(G,with_labels=True,font_weight='bold')
```

```python
nx.draw(H,with_labels=True,font_weight='bold')
```

```python
#通过列表添加边
G.add_edges_from([(1,3)])
nx.draw(G,with_labels=True,font_weight='bold')
```

```python
#从边容器中添加边
G.add_edges_from(H.edges)
nx.draw(G,with_labels=True,font_weight='bold')
```

```python
#删除节点
G.remove_nodes_from([1,3])
list(G.nodes)
```

```python
#删除边
G.remove_edge(4,5)
nx.draw(G,with_labels=True,font_weight='bold')
```

```python
#清除
G.clear()
print(G.nodes)
```

```python
#节点的数据类型可以不一样
G.add_nodes_from([1,'a'])
G.add_edge(1,'a')
nx.draw(G,with_labels=True,font_weight='bold')
```

```python
#无向图也有邻居（类似邻接矩阵）
G.add_edge(1,'b')
print(G.adj[1])
print(G.degree[1])
print(G[1])
print(G[1]['a'])
```

```python
#遍历无向图并设置边的权重（无向图会遍历2次）
for n,nbrs in G.adj.items():#每个节点，以及该节点的邻接边字典
    for nbr,eattr in nbrs.items():#每个邻接边的节点以及对应的属性字典
        eattr['w']=1
        print(nbr)
```

```python
#图可以有属性，节点可以有，边也可以有
G.graph['day']='Friday'
G.graph
```

```python
#节点属性
G.nodes[1]['room']=714
G.nodes['a']['room']=234
G.nodes.data()
```

```python
#边的属性
print(G[1]['a']['w'])
print(G.edges[1,'a']['w'])
```

```python
#有向图
DG = nx.DiGraph()
DG.add_edge(1,2)
DG.add_edge(1,3)
DG.add_edge(2,3)
nx.draw(DG,with_labels=True,font_weight='bold')
```

```python
#有向图的边是由前驱和后继的
print(list(DG.successors(2)))
print(list(DG.predecessors(2)))
```
